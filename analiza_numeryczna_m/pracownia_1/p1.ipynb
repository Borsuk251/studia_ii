```julia
function my_sin(x)#wlasna implementacja sinusa z taylora
    sign = 1.0
    if x<0
        x=-x
        sign = -sign
    end
    while x >pi #|| x < -pi#ograniczeni (-pi,pi)
            x-=2*pi
    end
    if x>pi
        x-=pi
        sign = -sign   
    end
    if x<-pi
        x+=pi
        sign = -sign
    end
    if  x < -0.5*pi #|| x > 0.5*pi #ograniczenie (-pi/2,pi/2)
        #if x>0
            x+=pi
            sign = -sign
    end
    if 0.5*pi < x
        x-=pi
        sign = -sign
        #end
    end
     res = x*(1-x*x/(2*3)*(1-x*x/(4*5)*(1-x*x/(6*7)*(1-x*x/(8*9)*
                    (1-x*x/(10*11)*(1-x*x/(12*13)*(1-x*x/(14*15)*(1-x*x/(16*17)))))))))
    return res*sign#uwzgledniamy znak
end
```

```julia
function my_half(x, n)#n krokow
    while x > pi
        x-=2*pi
    end
    while x<-pi
        x+=2*pi
    end
    sign_c, sign_s, check = 1, 1, 0
    if x>pi
        x-=pi
        sign_c, sign_s = -sign_c, -sign_s   
    end
    if x<-pi
        x+=pi
        sign_c, sign_s = -sign_c, -sign_s
    end
    if  x < -0.5*pi || x > 0.5*pi
        if x>0
            x=pi-x
            sign_c = -1
        else
            x+=pi
            sign_c, sign_s = -sign_c, -sign_s
        end
    end    
    if x<0 && x>=-pi/2
       x+=pi/2 
       check = 1 
    end
    cos2, sin2 = 0, 1 #wartosci w pi/2
    res, angle_t = 0,0
    y = pi/2#kolejne potegi 2 do minus razy pi
    sin1, cos1 = 1, 0 #od pi/2 bo musimy dzielic katy
    c3,s3 = 0,0
    cos3,sin3 = 1,0 #wartosci w zerze
    for i = 1:n
        sin1, cos1 = sin2, cos2
        y *= 0.5#nowy kat np pi/4
        angle_t += y
        cos2 = sqrt((1+cos1)/2)#dla kata polowkowego pi/4
        sin2 = sqrt((1-cos1)/2)#teraz dla pi/4
        if x > angle_t #jesli kat wiekszy od pi/4
            c3 = cos3*cos2-sin3*sin2 #nowe katy dla A+B
            s3 = sin3*cos2 + cos3*sin2# czyli tez nowe AAAAA
            cos3 ,sin3 = c3, s3
        else
            angle_t -= y #jesli chcemy mniejsza potege, bedziemy dodawac kolejna mniejsza
        end  
     end
    if check ==1
        return [my_half(-x+pi/2,n)[1],-cos3]
    end
    return [cos3*sign_c,sin3*sign_s]
end
```
