```python
def dl1(x):
    res = 0
    l = len(x)
    for i in range(l):
        res+= x[i][0]**2
    return math.sqrt(res)
```

- długość wektora x,
- średnią ważoną wektora x z wagami w,
- odległość euklidesową między wektorami x i y,

```python
def distance0(x):
    return np.sqrt(np.sum(x**2))
def mean_w(x,w):
    return np.sum(x*w)/np.sum(w)#np.dot(x.T,w)/np.sum(w) #wzraca tablice
def dist_vect(x,y):
    return distance0(x-y)
def scalar_pr(x,y):
    return np.sum(x*y)#np.dot(x.T,y) #np.dot-> matrix mul
```

b) Niech X będzie macierzą ustalonego rozmiaru d x N zawierającą N wektorów kolumnowych
długości d. Niech y i w będą wektorami kolumnowymi długości d. Policz
- długości kolejnych wektorów z macierzy X (wyznacz wektor długości N zawierający te
długości),<br>
-średnią ważoną kolejnych wektorów z macierzy X z wagami w (wyznacz wektor długości N
zawierający te średnie),
- odległości euklidesowe między kolejnymi wektorami z macierzy X i wektorem y (wyznacz
wektor długości N zawierający te odległości),
- iloczyny skalarne kolejnych wektorów z macierzy X i wektora y (wyznacz wektor długości N
zawierający te iloczyny).
Obliczenia przeprowadź dla losowo wygenerowanej macierzy X i losowo wygenerowanych
wektorów y i w (dla d = 100 i N = 1000).

```python
d2=100;N=1000
X1 = np.random.rand(d2,N)
y1 = np.random.rand(d2,1)
w1 = np.random.rand(d2,1)
```

```python
def dist_matrix(x):
    return np.sqrt(np.sum(x**2, axis=0)) #suma po kolumnach, np.square

def mean_w(x,w): #wrsja macierzowa
    return np.sum(x*w)/np.sum(w)#broadcasting, mozna tez np.sum(np.dot(x.T,w))/...
def m_w(x,w,N): #wersja z forem, mniej wydajna
    mw = np.zeros((1,N)) 
    sw = np.sum(w)
    for i in range(N): #kolejne srednie wazone
        mw[0][i] = np.sum(np.dot(x[:,i].T,w))#np.sum(x[:,i]*w) tak nie mozna bo braodacgsing i macierz prostokatna 
    return np.sum(mw)/np.sum(w)

def dist_euc(x,y): #odl miedzy wekt
    return np.sqrt(np.sum((x-y)**2, axis=0))
def eu2(X,y,N):
    dl_e = np.zeros((1,N)) 
    l = np.shape(X)[0]
    X_new = X-y
    for i in range(N):
        dl_e[0][i] = dist_matrix(X_new[:,i])
    return dl_e

def scal_pr(x,y):#iloczyn skalarny
    return np.sum(x*y, axis=0)
def s2(X,y,N):
    il_sk = np.zeros((1,N))
    for i in range(N):
        il_sk[0][i] = X[:,i].T.dot(y)
    return np.sum(il_sk, axis=0)
```

```python
print(dist_matrix(X1),scal_pr(X1,y1), dist_euc(X1,y1), mean_w(X1,w1))
```
