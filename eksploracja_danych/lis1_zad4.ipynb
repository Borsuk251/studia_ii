```python
import matplotlib.cm as cm
from matplotlib.colors import ListedColormap
sepal_length = iris.data[:100,0]
sepal_width = iris.data[:100,1]
iris_y = np.array(iris.target[:100]) #podzial na gatunki
#print(iris_y)
m = 'flag' #gotowe colormap
m2 = ListedColormap(['#FF0000', '#0000FF', '#00FF00']) #RGB
z = np.arange(4,9)
#l = np.linspace(np.min(sepal_length), np.max(sepal_length))

plt.scatter(sepal_length, sepal_width, c=iris_y, cmap=m2,edgecolors='black')#c->dzielenie na kolory
plt.plot(z,2*z-8, c='blue') #dodajemy prosta
plt.title('divide2')
plt.xlim(4,7.5)
plt.ylim(1.5,5)
plt.savefig('zadanie4d.eps')
plt.show()

print(iris_y)
iris_y = [iris_y[i]^1 if 2 * sepal_length[i] - 8 < sepal_width[i] 
          else iris_y[i] for i in range(len(iris_y))] #zamiana pod prosta
#for i in range(len(iris_y)):
 #   if 2 * sepal_length[i] - 8 < sepal_width[i]:
  #      iris_y[i]+=1
print(iris_y)
plt.scatter(sepal_length, sepal_width, c=iris_y, cmap=m2)#c->dzielenie na kolory
plt.plot(z,2*z-8, c='blue')
plt.title('z podana prostą')
plt.xlim(4,7.5)
plt.ylim(1.5,5)
plt.savefig('zadanie4f.eps')
plt.show()
#wykres z przesunieta prosta
plt.scatter(sepal_length, sepal_width, c=iris_y, cmap=m2)#c->dzielenie na kolory
plt.xlim(4,7.5)
plt.ylim(1.5,5)
plt.title('zmiana prostej')
plt.plot(z,2*z-8, c='blue')
plt.plot(z,1.50*z-4.9, c='red')#zmiana wspolczynnikow prostej
plt.savefig('zadanie4ff.eps')
plt.show()
```
