```python
import random
import matplotlib.pyplot as plt
%matplotlib inline
params = {"ytick.color" : "w",
          "xtick.color" : "w",
          "axes.labelcolor" : "w",
          "axes.edgecolor" : "w",
         "text.color" : "w"}
plt.rcParams.update(params) #dark theme
import scipy.stats
import scipy
import matplotlib.colors as mcolors
from matplotlib import cm
```

```python
def cloud(a,b,n):
    mean = [a,b]
    cov = np.array([[1,0],[0,1]])
    return np.random.multivariate_normal(mean, cov, n).T
def gen(n,x):#n chmur dla wektora X wspol
    return np.array([cloud(a,b,n) for (a,b) in x])

n=1000
x = np.array([[5,5],[5,15],[30,20]])
#print(len(x))
points = gen(n,x)
#print(np.shape(points))
w1 = plt.figure().gca()
for i in points:
    w1.scatter(i[0],i[1],s=3.5) #s wielkosc pkt
plt.show()
w2 = plt.figure().gca()
x1 = points[:,0,:].reshape(n*len(x)) #wektor wspolrzednych
y1 = points[:,1,:].reshape(n*len(x))
h, xedges, yedges, image = plt.hist2d(x1, y1, bins=100)#histagram pkt
```

```python
def vect(K,d): #tworzenie wektorow wspolrzednyhc wierzcholkow
    r = d/(2*math.sin(math.pi/K))#promien okregu
    A = np.zeros((K,2)) #macierz do wspol wierzcholkow
    k=0
    for i in np.arange(0,2*math.pi,2*math.pi/K): #zakres
        A[k,:] = [r*math.sin(i), r*math.cos(i)] #kolejne wierzcholki
        k+=1 #calkowite indeksy
    return A

a1 = vect(7,5)
a2 = vect(11,10)
a3 = vect(23,15)
#print(a1)
b1 = gen(n,a1)
b2 = gen(n,a2)
b3 = gen(n,a3)

y1 = plt.figure().gca()
for i in b1:
    y1.scatter(i[0],i[1],s=3)#same punkty
plt.show()
#reshape do wektora
h, xedges, yedges, image = plt.hist2d(b1[:,0,:].reshape(n*len(b1)), b1[:,1,:].reshape(n*len(b1)), bins=100, density=True)
y2 = plt.figure().gca()
for i in b2:
    y2.scatter(i[0],i[1],s=3)
plt.show()
h, xedges, yedges, image = plt.hist2d(b2[:,0,:].reshape(n*len(b2)), b2[:,1,:].reshape(n*len(b2)), bins=100, density=True)
y3 = plt.figure().gca()
for i in b3:
    y3.scatter(i[0],i[1],s=3)
plt.show()
h, xedges, yedges, image = plt.hist2d(b3[:,0,:].reshape(n*len(b3)), b3[:,1,:].reshape(n*len(b3)), bins=100, density=True)
```
