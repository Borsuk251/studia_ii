```
def create_bigrams():
  dic = {}
  for bi in letters21:
    if bi[0] in dic:
      dic[bi[0]].append((letters21[bi],bi[1]))
    else:
      dic[bi[0]] = [(letters21[bi],bi[1])]

  for first in dic:
    sum = 0
    for values in dic[first]:
      sum+=values[0]
    for i in range(len(dic[first])):
      dic[first][i] = (dic[first][i][0]/sum, dic[first][i][1])
    
  return dic
dic_biletters = create_bigrams()
```

```
def ppmi(word, unigrams, bigrams, k=10):
  results = {}
  LEN_UNI = len(unigrams)
  LEN_BI = len(bigrams)
  res_words = []
  res_num = []

  for elements in bigrams[word]:
   # print(elements)
    log = np.log((elements[0]) / (unigrams[word] * unigrams[elements[1]])*LEN_UNI**2 /LEN_BI)
    #print(log)
    res_words.append(elements[1])
    res_num.append(log)
  res_num = (res_num-min(res_num))/(max(res_num)-min(res_num))
  for i in range(len(res_num)):
    results[word+res_words[i]]= res_num[i]
#  results = sorted(results, key=itemgetter(1), reverse=True)
  return results
ppmi('a', letters, dic_biletters)

def ppmi_dic():
  ppmi_dic = {}
  for letter in letters:
    temp_dic = ppmi(letter, letters, dic_biletters)
    for bi in temp_dic:
      ppmi_dic[bi] = temp_dic[bi]
  return ppmi_dic
ppmi_dic = ppmi_dic()
```
