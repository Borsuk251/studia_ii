```
import re
def load_rhytm_sent():
  with open('/content/drive/My Drive/nlp/rytmiczne_zdania_z_korpusu.txt', 'r', encoding = 'utf8') as file:
    corpus = []
    for line in file:
      sent = line.split()
      if line[0]=='P':
        tmp = [i for i in re.split('POTENCJALNY_RYM:|\[\*\]', line) if i!=''] #smart regex
      if line[0] == 'R':
        tmp = [i for i in re.split('RYM:|\[\*\]', line) if i!='']
      tmp =[j.split() for j in tmp]
      corpus.append(tmp)
  return corpus
corpus = load_rhytm_sent()
```

```
def load_supertags():
  tags, words = {}, {}   
  with open('/content/drive/My Drive/nlp/supertags.txt', 'r', encoding = 'utf8') as file:
    for line in file:
      line = line.split()
      words[line[0]] = line[1]
      if line[1] not in tags:
        tags[line[1]] = [line[0]]
      else:
        tags[line[1]].append(line[0])
  return tags, words
tags, words = load_supertags()
```

```
import numpy as np
def load_vectors():
  vectors = {}
  with open('/content/drive/My Drive/nlp/poleval_base_vectors.txt', 'r', encoding = 'utf8') as file:
    for i, line in enumerate(file):
      line = line.split()
      num = np.array([float(i) for i in line[1:]]) #way for loading vectors
      vectors[line[0]] = num
      #if i==10:break
  return vectors
vectors = load_vectors()
```

```
def load_parts(): #wczytanie części mowy
  parts = {}
  with open('/content/drive/My Drive/nlp/skladnicaTagsBases.pl', 'r', encoding = 'utf8') as file:
    for i, line in enumerate(file):
      line = line.split()

      part = line[-1].split(':')[0]
      if part[-1] == '.': part = part[:-2]
      if line[0][12:-2] in parts:
        parts[line[0][12:-2]].append(part)
      else:
        parts[line[0][12:-2]] = [part]
  return parts
parts = load_parts()
```

```
def check_change(word): #sprawdzamy, czy możemy zmienić, zmieniamy dla rzeczowników, czasowników, przymiotników, przysłówków
  if word not in parts:
    return False
  parts_change = {'subst', 'depr', 'adj', 'adja', 'adjp', 'adjc', 'pact', 'adv','fin', 'bedzie', 'praet', 'impt', 'imps', 'pcon', 'pant', 'ger', 'pact', 'ppas'}
  for p in parts[word]:
    if p in parts_change:
      return True
  return False
```

```
vowels = ['a', 'ą','e', 'ę', 'i', 'o', 'ó', 'u', 'y']

def count_vowels(sent): #count syllabs in sent (array) for each word
  ans = []
  for word in sent:
    count = 0
    for i in range(len(word)):
      if word[i] in vowels:
        count += 1
        if i>0 and word[i-1] == 'i':
          count-=1
    ans.append(count)
  return ans

def get_rhym(word): #get posssible rhyme
  count = 0
  if word[-1] in vowels:
    count +=1
  for i in range(len(word)-2,0,-1):
    if word[i] in vowels:
      count +=1
    if word[i] == 'i' and word[i+1] in vowels:
      count -=1
    if count ==2:
      return word[i:]
  return word
```

```
def find_most_similar(words, word2,  same_syllabs = False):
  max = 0
  tab = []
  word2_base, best = '', ''
  if word2 in base:
    word2_base = base[word2] # get base form
  else:
    return False
  for word in words:
    word_base = ''
    if word in base:
      word_base = base[word]
    else:
      continue
    if same_syllabs:
      if count_vowels(word) != count_vowels(word2):
        continue
    if word_base in vectors and word2_base in vectors:

      new_val = np.dot(vectors[word_base], vectors[word2_base]) #measure how close
      
      if new_val > max and word2_base != word_base: #get closest
        best = word
        max = new_val
  if best != '':
    return best
  return False
```

```
def find_best_rhym(word, tag2, word2, same_syllabs = False):
  possible = []
  for i in tags[tag2]:
    if count_vowels(i) == count_vowels(word) and get_rhym(word) == get_rhym(i) and i!=word:
      possible.append(i)
  return find_most_similar(possible, word2, same_syllabs) #find most similar word for word2
```

```
import copy
import random

def gen_poem(all=False, same_syllabs = False):
  sent = corpus[np.random.choice(np.arange(len(corpus)))] #get random sent from corpus
  new_sent = copy.deepcopy(sent)
  ponct = [',', '.', '!', '?']
  id1_rhym = 0
  id2_rhym = 0
  for w in np.arange(len(sent[0])-1,0, -1): #get id of last word sent1
    if sent[0][w] not in ponct:
      id1_rhym =w
      break
  for w in np.arange(len(sent[1])-1,0, -1): #get id of last word sent2
    if sent[1][w] not in ponct:
      id2_rhym =w
      break
  
  first = sent[0][id1_rhym] #first word to rhyme
  if sent[1][id2_rhym] not in words: #if can't find tag for word get new sent
    return gen_poem(all, same_syllabs)

  #find closest word (vector) with same syllabs length tag and rhyme
  second = find_best_rhym(first, words[sent[1][id2_rhym]], sent[1][id2_rhym], same_syllabs)


  if not second: #if close word not found, search for it
    for i in range(100):
      if second: #if found
        break
      tmp = find_most_similar(tags[words[first]], first, same_syllabs)# find similar word in meaning, with other word1
      possible = tags[words[first]]
      first = possible[np.random.choice(np.arange(len(possible)))]
      if tmp:
        tmp = first #change first word to rhyme
      second = find_best_rhym(first, words[sent[1][id2_rhym]], sent[1][id2_rhym], same_syllabs)
    if not second:
      return gen_poem(all, same_syllabs ) # if failed start all over again

  new_sent[0][id1_rhym] = first
  new_sent[1][id2_rhym] = second
  if all:
    for i in range(id1_rhym):
      #added probability of changing, if same syllabs required always search for new word (computation time)
      if random.random() > 0.5 or same_syllabs:  
        if sent[0][i] not in ponct and check_change(sent[0][i]): #check if part of speech is correct
          w2 = find_most_similar(tags[words[sent[0][i]]], sent[0][i],same_syllabs)
          if w2:
            new_sent[0][i] = w2 #change word
      if random.random() > 0.5 or same_syllabs:  
        if sent[1][i] not in ponct and check_change(sent[1][i]):
          w2 = find_most_similar(tags[words[sent[1][i]]], sent[1][i],same_syllabs)
          if w2:
            new_sent[1][i] = w2
  return sent, new_sent
```

```
def printing(sent1,sent2):
  print(' '.join(sent1[0]))
  print(' '.join(sent1[1]))
  print(' '.join(sent2[0]))
  print(' '.join(sent2[1]))
```
