```
def create_bigrams():
  bigrams_dict = {}
  with open('/content/drive/My Drive/nlp/poleval_2grams.txt', 'r', encoding = 'utf8') as file:
    for i, line in enumerate(file):
      #if i >= 10000000:
      #  break
      #print(line)
      words = line.split()
      how_many, word1, word2 = int(words[0]), words[1], words[2]  #cast to int
      #print(words)
      if how_many < 5:
        continue
      else:
        if word1 in bigrams_dict:
          bigrams_dict[word1].append((how_many, word2))
        else:
          bigrams_dict[word1] = [(how_many, word2)]
  return bigrams_dict
bigrams_dict = create_bigrams()
#print(bigrams_dict)
```

```
def create_trigrams():
  trigrams_dict = {}
  with open('/content/drive/My Drive/nlp/poleval_3grams.txt', 'r', encoding = 'utf8') as file:
    for i, line in enumerate(file):
      if i >= 10000000:
        break
      #print(line)
      words = line.split()
      #print(words)
      if words[2] == '<EOS>':
        continue
      else:
        how_many, word1, word2, word3 = int(words[0]), words[1], words[2], words[3] #cast to int
      #print(words)
        if how_many < 5:
          continue
        else:
          if (word1,word2) in trigrams_dict: 
            trigrams_dict[(word1, word2)].append((how_many, word3))
          else:
            trigrams_dict[(word1, word2)] = [(how_many, word3)]
  return trigrams_dict
trigrams_dict = create_trigrams()
#print(trigrams_dict)
```

```
def text_generation_bigrams(bigrams_dict, n_words):
  first_word = random.choice(list(bigrams_dict.keys()))
  text = []
  #str_text = ''
  text.append(first_word)
  for i in range(n_words):
    if first_word in bigrams_dict.keys():
      value_elements = len(bigrams_dict[first_word])
      #print(value_elements)
      chosen_word = random.randint(0, value_elements)
      new = bigrams_dict[first_word][chosen_word-1][1]
      text.append(new)
      first_word = new
    else:
      first_word = random.choice(list(bigrams_dict.keys()))
      text.append(' STOP ')
      text.append(first_word)
  print(' '.join(text))
  return text
```

```
def text_generation_trigrams(trigrams_dict, n_words):
  first_word = random.choice(list(trigrams_dict.keys()))
  text = []
  #str_text = ''
  text.append(first_word)
  end = False
  for i in range(n_words):

    if first_word in trigrams_dict.keys():
      value_elements = len(trigrams_dict[first_word])
      #print(value_elements)
      chosen_word = random.randint(0, value_elements)
      
      new = trigrams_dict[first_word][chosen_word-1][1]
 
      text.append(new)
      first_word = (first_word[1], new)
    else:
      w = [j for j in trigrams_dict.keys() if j[0]==first_word[1]]

      if len(w) == 0:
        first_word = random.choice(list(trigrams_dict.keys()))
        text.append(first_word)
        continue

      chose_with_first = random.randint(0,len(w))
      #print(w, len(w), chose_with_first)
      len_new2 = len(trigrams_dict[(first_word[1], w[chose_with_first-1][1])])
      new2_id = random.randint(0,len_new2)
      new2 = trigrams_dict[(first_word[1], w[chose_with_first-1][1])][new2_id-1][1]

      text.append(new2)
      first_word = (w[chose_with_first-1],new2)

  #print(' '.join(text))
  return text
```

```
def normalization_dict(values):
  frequency = []
  for x in values:
    frequency.append(int(x[0]))
  sum_all = sum(frequency)
  return [i/sum_all for i in frequency]
```

```
def text_generation_bigrams_with_prop(bigrams_dict, n_words):
  first_word = random.choice(list(bigrams_dict.keys()))
  text = []

  text.append(first_word)
  for i in range(n_words):
    if first_word in bigrams_dict.keys():
      value_elements = len(bigrams_dict[first_word])
      
      weights = normalization_dict(bigrams_dict[first_word])
      chosen_word = np.random.choice(value_elements, p=weights)
      #print(value_elements)
      #chosen_word = random.randint(0, value_elements)
      new = bigrams_dict[first_word][chosen_word-1][1]
      text.append(new)
      first_word = new
    else:
      first_word = random.choice(list(bigrams_dict.keys()))
      text.append(' STOP ')
      text.append(first_word)
  print(' '.join(text))
  return text
```

```
def text_generation_trigrams_with_prop(trigrams_dict, n_words):
  first_word = random.choice(list(trigrams_dict.keys()))
  text = []
  #str_text = ''
  text.append(first_word)
  end = False
  for i in range(n_words):

    if first_word in trigrams_dict.keys(): # two word key
      value_elements = len(trigrams_dict[first_word])
      #print(value_elements)
      weights = normalization_dict(trigrams_dict[first_word])
      chosen_word = np.random.choice(value_elements, p=weights) # randomizing with weights (distribution of values)
      #chosen_word = random.randint(0, value_elements)
      
      new = trigrams_dict[first_word][chosen_word-1][1]
 
      text.append(new)
      first_word = (first_word[1], new)
    else: # case when searching with one word key
      w = [j for j in trigrams_dict.keys() if j[0]==first_word[1]] 

      if len(w) == 0: #search for new tuple
        first_word = random.choice(list(trigrams_dict.keys()))
        text.append(first_word)
        continue
      
      chose_with_first = random.randint(0,len(w)) #choosing second part of key
      #print(w, len(w), chose_with_first)

      len_new2 = len(trigrams_dict[(first_word[1], w[chose_with_first-1][1])])

      weights2 = normalization_dict(trigrams_dict[(first_word[1], w[chose_with_first-1][1])])
      new2_id = np.random.choice(len_new2, p=weights2)
      #new2_id = random.randint(0,len_new2)
      new2 = trigrams_dict[(first_word[1], w[chose_with_first-1][1])][new2_id-1][1]

      text.append(new2)
      first_word = (w[chose_with_first-1],new2)

  #print(' '.join(text))
  return text
```

```
def measure_how_natural(sentence, dict):
  sentence = sentence.lower()
  sentence = sentence.split()

  to_perm = [s.translate(str.maketrans('', '', string.punctuation)) for s in sentence]
  permutations_object = itertools.permutations(to_perm)
  permutations_list = list(permutations_object)
  print(len(permutations_list))
  score = 0
  perm_list_with_scores = {}
  for perm in permutations_list:
    #print(perm)
    score = 0 
    for word1, word2 in zip(perm, perm[1:]):
      #print(word1,word2)
      if word1 in dict:
        #values_len = len(bigrams_dict[word1])
        for values in dict[word1]:
          if values[1] == word2:
            score+= int(values[0])
    #print(score)

    perm_list_with_scores[perm] = score
  new_dict = {k: v for k, v in sorted(perm_list_with_scores.items(), key=lambda item: item[1], reverse=True)}
  #print(perm_list_with_scores)
  n_key = 0
  for key in new_dict:
    print(key, new_dict[key])
    n_key+=1
    if n_key > 20:
      break

measure_how_natural('Judyta dała wczoraj Stefanowi czekoladki.', bigrams_dict_sent)
measure_how_natural('Babuleńka miała dwa rogate koziołki', bigrams_dict_sent)
measure_how_natural('Wczoraj wieczorem spotkałem pewną piękną kobietę', bigrams_dict_sent)

measure_how_natural('wypiłem kawę, herbatę i mleko', bigrams_dict_sent)
measure_how_natural('dzisiaj będziemy mieć dłuższe zajęcia', bigrams_dict_sent)
measure_how_natural('zostałem wyznaczony przez Odyna', bigrams_dict_sent)
```

```
def create_bigrams_ends():
  bigrams_dict = {} #{ word1 : (count, word2)}
  bigrams_ends = {} #{(word1_end, word2_end) : count}
  with open('/content/drive/My Drive/nlp/poleval_2grams.txt', 'r', encoding = 'utf8') as file:
    for i, line in enumerate(file):
      #if i >= 10000000:
      #  break
      #print(line)
      words = line.split()
      how_many, word1, word2 = int(words[0]), words[1][-3:], words[2][-3:]  #cast to int
      #print(words)
      if how_many < 5:
        continue
      if (word1, word2) in bigrams_ends:
        num = bigrams_ends[(word1, word2)]
        bigrams_ends[(word1, word2)] = num+how_many
      else:
        bigrams_ends[(word1,word2)] = how_many

  for words in bigrams_ends:
    if words[0] in bigrams_dict:
      bigrams_dict[words[0]].append((bigrams_ends[words], words[1]))
    else:
      bigrams_dict[words[0]] = [(bigrams_ends[words], words[1])]

  return bigrams_dict, bigrams_ends
bigrams5, bigrams5_ends = create_bigrams_ends()
#print(bigrams_dict)
```
